Escreva um programa que forneça o tipo de aplicação financeira adequado a um investidor a partir de dois dados fornecidos: o grau de aceitação de risco e o valor a ser investido. O quadro 4.4 mostra as opções de combinação entre esses dois dados. O grau de aceitação de risco deve ser lido do teclado na forma BX para baixo ou AL para alto. Se for fornecido algo diferente disso o programa deve mostrar uma mensagem indicando que foi fornecido dado inválido. Para o valor deve-se ler um número real.<br>
| Aceitação de risco | Valor < 1000,00 | Valor >= 1000,00 |
| :---: | :---: | :---: |
|Baixo (BX) | Poupança | Renda fixa |
|Alto (AL) | Bitcoins | Ações |

```python
risco = input('Digite BX ou AL para o grau de risco: ')
valor = float(input('Digite o valor: '))
if risco != 'BX' and risco != 'AL':
  print(f'{risco} é inválido para o grau de risco.')
else:
  if risco == 'BX':
    if valor < 1000.0:
      tipo = 'Poupança'
    else:
      tipo = 'Renda fixa'
  else: # risco == 'AL"
    if valor < 1000.0:
      tipo = 'Bitcoins'
    else:
      tipo = 'Ações'
  print(f'Você deve investir em {tipo}.')
```

**Exercício Proposto 4.10**<br>
Escreva um programa que leia 3 números inteiros e mostre na tela se eles formam um triângulo ou não. Caso formem um triângulo informe o tipo de triângulo (equilátero, isósceles ou escaleno).<br>
Regra: Para três números formarem um triângulo precisa ocorrer que:<br>
a) os três números precisam ser maiores que zero;<br>
b) a soma dos dois menores valores deve ser maior que o terceiro.

```python
A = int(input('Digite o valor do lado A: '))
B = int(input('Digite o valor do lado B: '))
C = int(input('Digite o valor do lado C: '))
if A > 0 and B > 0 and C > 0:
  if A + B > C and B + C > A and A + C > B:
    if A == B == C:
      print(f'Os lados {A}, {B} e {C} formam um triângulo equilátero.')
    elif A == B != C or A != B == C:
      print(f'Os lados {A}, {B} e {C} formam um triângulo isósceles.')
    else:
      print(f'Os lados {A}, {B} e {C} formam um triângulo escaleno.')
  else:
    print(f'Os lados {A}, {B} e {C} não formam um triângulo.')
else:
  print('Há pelo menos um valor inválido.')
print('Fim do programa.')
```

**Exercício Proposto 4.13**<br>
Nas eleições municipais os municípios com 200 mil eleitores ou mais tem segundo turno caso o primeiro colocado não tenha mais do que 50% dos votos. Escreva um programa que leia o nome do município, a quantidade de eleitores e a quantidade de votos do candidato mais votado e informe se haverá segundo turno ou não.

```python
municipio = input('Digite o nome do município: ')
eleitores = int(input('Digite a quantidade de eleitores: '))
votos = int(input('Digite a quantidade votos do candidato mais votado: '))
if eleitores < 200000:
  print(f'O município {municipio} não tem segundo turno.')
else:
  if eleitores / 2 < votos:
    print(f'O município {municipio} tem segundo turno.')
  else:
    print(f'O município {municipio} não tem segundo turno.')
print('Fim do programa.')
```

**Exercício Proposto 4.14**<br>
Em um determinado momento do dia a cotação de compra das moedas estrangeiras é a seguinte:<br>
Dólar: US$ 1.00 = R$ 4.89<br>
Euro: € 1.00 = R$ 5.26<br>
Libra Esterlina: £ 1.00 = R$ 6.17<br>
Escreva um programa que leia o tipo (D, E ou L maiúsculo) e o valor de moeda estrangeira que se quer comprar e calcule o valor em reais necessários.

```python
moeda = input('Digite o tipo da moeda (D, E ou L): ')
valor = float(input('Digite o valor da moeda: '))
if moeda == 'D':
  reais = valor * 4.89
  print(f'Para comprar {valor:.2f} dólares são necessários R$ {reais:.2f}.')
elif moeda == 'E':
  reais = valor * 5.26
  print(f'Para comprar {valor:.2f} euros são necessários R$ {reais:.2f}.')
elif moeda == 'L':
  reais = valor * 6.17
  print(f'Para comprar {valor:.2f} libras esterlinas são necessários R$ {reais:.2f}.')
else:
  print(f'Tipo {moeda} inválido.')
print('Fim do programa.')
```
